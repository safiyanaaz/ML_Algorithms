```python
import numpy as np
import pandas as pd
df=pd.read_csv(r'C:\Users\Lenovo\Desktop\ML course\Deep learning\Financial Distress.csv')
```

```python
df.head()
```

```python
#The company which have greater value than -0.5 are healthy value else they are financial distress company classification
df['fd']=pd.np.where(df['Financial Distress'] > -0.50,0,1)
```

```python
df.describe()
```

```python
#changing data types
columns=['Company','Time','x80','fd']
for i in columns:
    df[i]=df[i].astype(str)
```

```python
df.info()
```

```python
df=df.drop('Financial Distress', axis=1)
```

```python
#taking percentage of null values
df.isnull().sum()/df.shape[0]*100
```

```python
X=df.drop('fd',axis=1)
```

```python
Y=df['fd']
```

```python
Y.value_counts()
```

```python
X=pd.get_dummies(X)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.decomposition import PCA
```

```python
#most important features
pca=PCA(whiten=True)
pca.fit(X_train)
variance=pd.DataFrame(pca.explained_variance_ratio_)
np.cumsum(pca.explained_variance_ratio_)
```

```python
pca=PCA(n_components=15,whiten=True)
pca_X_train=pca.fit_transform(X_train)
pca_X_test = pca.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(pca_X_train)
X_test=sc.fit_transform(pca_X_test)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp=MLPClassifier(solver='lbfgs' , alpha =1e-05,hidden_layer_sizes=(5,2),random_state=1)
```

```python
mlp.fit(X_train,Y_train)
```

```python
y_pred = mlp.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_pred)
```
