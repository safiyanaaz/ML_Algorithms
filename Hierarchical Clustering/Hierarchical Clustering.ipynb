```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset=pd.read_csv(r'C:\Users\Lenovo\Desktop\Git Repos\ML_Algorithms\Hierarchical Clustering\Mall_Customers.csv')
```

```python
dataset.head()
#This dataset is of the Mall where Customer data is given
#The Mall wants based on Annual Income and Spending Score to see which person is spending more
```

```python
X=dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
#To know how many clusters must be formed
import scipy.cluster.hierarchy as sch
dendogram=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distance')
plt.show()
#suggesting us 3
```

```python
'''For more Accuracy Another method for finding max no of cluster
Find the biggest vertical line in the dendogram which does not cut by any horizontal line
thus green line of between 200-250 therefor max 5 clusters'''
```

```python
#Create a model to fit the Hierarchical modelling.
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(X) #on which cluster which datapoints lies
```

```python
#Visualizing the clusters
plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=100,c='red',label='Cluster-1')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=100,c='blue',label='Cluster-2')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=100,c='green',label='Cluster-3')
plt.scatter(X[y_hc==3,0],X[y_hc==3,1],s=100,c='cyan',label='Cluster-4')
plt.scatter(X[y_hc==4,0],X[y_hc==4,1],s=100,c='magenta',label='Cluster-5')
plt.title('Clusters of the Customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
