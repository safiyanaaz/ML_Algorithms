```python
import matplotlib.pyplot as plt
%matplotlib inline


plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()

'''You may be wondering why the x-axis ranges from 0-3 and the y-axis from 1-4. 
If you provide a single list or array to the plot() command, matplotlib assumes it is a 
sequence of y values, and automatically generates the x values for you. Since 
python ranges start with 0, the default x vector has the same length as y but starts with 0. Hence the x data are [0,1,2,3].

'''
```

```python
#plt.plot(x,y)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
'''The axis() command in the example above takes a list of [xmin, xmax, ymin, ymax] 
and specifies the viewport of the axes.'''
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.xlabel('numbers')
plt.ylabel('another_numbers')
plt.title('Practice')
plt.show()
```

```python
#multiple plots
#plt.subplot(nrows,ncols,index)
plt.subplot(2,3,1)
plt.plot(x,y,'ro')
plt.subplot(2,3,2)
plt.plot(x,y,'g^')

plt.subplot(2,3,3)
plt.plot(x,y,'r^')

plt.subplot(2,3,4)
plt.plot(x,y,'b--')


plt.subplot(2,3,6)
plt.plot(x,y,'m*')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])      #[xmin,ymin,xmax,ymax]
axes1.plot(x,y)
axes1.set_xlabel('x1',)
axes1.set_ylabel('x2')
axes1.set_title('practice')

axes2=fig.add_axes([0,0,0.5,0.5])
axes2.plot(x,y,'ro')

axes1.set_xlabel('new_x1',)
axes1.set_ylabel('new_x2')
axes1.set_title('practice_2')
```
