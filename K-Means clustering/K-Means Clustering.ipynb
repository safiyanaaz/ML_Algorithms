# K-Means Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans

#importing dataset

dataset =pd.read_csv(r"C:\Users\Lenovo\Desktop\Git Repos\ML_Algorithms\K-Means clustering\Churn_Modelling.csv")
dataset.head()
```

```python
#Based on Balance and Estimated Salary
x=dataset.iloc[:,[8,12]].values
```

```python
#Determining optimal  number of clusters
#keeping value of sum of squared errors in wcss array
wcss=[]
for i in range(1,16):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans++.inertia_)
    
    
    

```

```python
plt.plot(range(1,16),wcss)
plt.title('Elbow method')
plt.xlabel('No of Clusters')
plt.ylabel('WCSS')
plt.show()
#observing elbow method ,elbow is formed at 4 so optimal k=4 
```

```python
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=0)
y=kmeans.fit_predict((x))
```

```python
plt.scatter(x[y==0,0],x[y==0,1],s=25,c='red',label='Cluster 1')
plt.scatter(x[y==1,0],x[y==1,1],s=25,c='blue',label='Cluster 2')
plt.scatter(x[y==2,0],x[y==2,1],s=25,c='green',label='Cluster 3')
plt.scatter(x[y==3,0],x[y==3,1],s=25,c='cyan',label='Cluster 4')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=25,c='yellow',label='centroid')
plt.title('K-Means clustering')
plt.xlabel('Balance in account')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()

```
